```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
gender_counts = df['Sex'].value_counts()

plt.figure(figsize=(8, 5))
plt.bar(gender_counts.index, gender_counts.values, color=['blue', 'pink'])
plt.title('Gender Distribution on Titanic')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.show()

# Histogram for Age Distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Age'].dropna(), bins=20, edgecolor='black', color='green')
plt.title('Age Distribution on Titanic')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('train.csv')
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)

# Interactive Histogram for Age Distribution
fig = px.histogram(df, x='Age', nbins=20, title='Age Distribution on Titanic', color_discrete_sequence=['green'])
fig.update_layout(xaxis_title='Age', yaxis_title='Frequency', template='plotly_white')
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset
df = pd.read_csv('train.csv')
```

```python
# Handle missing data by filling NaN values with the median age
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
# Set the style of seaborn
sns.set(style="whitegrid")
```

```python
# Create a figure with subplots
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

# Advanced Bar Chart for Gender Distribution
gender_counts = df['Sex'].value_counts()
sns.barplot(x=gender_counts.index, y=gender_counts.values, hue=gender_counts.index,
            palette='pastel', ax=ax[0], dodge=False, legend=False)
ax[0].set_title('Gender Distribution on Titanic', fontsize=16)
ax[0].set_xlabel('Gender', fontsize=12)
ax[0].set_ylabel('Frequency', fontsize=12)
ax[0].bar_label(ax[0].containers[0], fmt='%d')  # Adding labels to bars

# Advanced Histogram for Age Distribution
sns.histplot(df['Age'], bins=20, kde=True, color='green', ax=ax[1])
ax[1].set_title('Age Distribution on Titanic', fontsize=16)
ax[1].set_xlabel('Age', fontsize=12)
ax[1].set_ylabel('Frequency', fontsize=12)

# Adding annotations to highlight median age
median_age = df['Age'].median()
ax[1].axvline(median_age, color='red', linestyle='--')
ax[1].text(median_age + 0.5, 40, f'Median Age: {median_age:.1f}', color='red', fontsize=12)

# Display the plots
plt.tight_layout()
plt.show()
```
